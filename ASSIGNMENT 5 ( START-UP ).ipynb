MODEL VALIDATION

THERE ARE TWO TYPES OF TECHINQUE:

     COLINEARITY CHECK 
    
     RESIDUAL ANALYSIS

```python
# 1. COLINEARITY CHECK
# VIF = 1/(1 - RSQUARE) OF EACH INDEPENDENT VARIABLES

rsq_rdspnd = smf.ols('RDSPND~ADMNSTRN+MRKTNGSPND',data = startup).fit().rsquared
vif_rdspnd = 1 / (1 - rsq_rdspnd)

rsq_admn = smf.ols('ADMNSTRN~RDSPND+MRKTNGSPND',data = startup).fit().rsquared
vif_admn = 1 / (1 - rsq_admn)

rsq_mrkspd = smf.ols('MRKTNGSPND~RDSPND+ADMNSTRN',data = startup).fit().rsquared
vif_mrkspd = 1 / (1 - rsq_mrkspd)

df = {'VARIABLES':["RDSPND","ADMNSTRN","MRKTNGSPND"],'VIF' : [vif_rdspnd,vif_admn,vif_mrkspd]}
VIF_DF = pd.DataFrame(df)
VIF_DF
```

```python
#none of the vif > 20 , so we can consider all the elements in the residual analysis since there is no colinearity
# 2. RESIDUAL ANALYSIS
# test for normality of model using qq plot
import statsmodels.api as sm
sm.qqplot(model.resid,line = 'q')
plt.title('NORMAL QQ PLOT OF RESIDUALS')
plt.show
```

```python
list(np.where(model.resid<-30000))
```

```python
# test for homoscedasticity or heteroscedasticity (standardized fittedvalues vs standardize residuals)
def standard_values(vals) : return (vals-vals.mean())/vals.std() #(z=x-)/sigma
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('RESIDUAL PLOT')
plt.xlabel('STANDARDIZED FITTED VALUES')
plt.ylabel('STANDARDIZED RESIDUAL VALUES')
plt.show()
```

```python
# test for errors
fig = plt.figure(figsize = (15,8))
sm.graphics.plot_regress_exog(model,'RDSPND',fig = fig)
```

```python
fig = plt.figure(figsize = (15,8))
sm.graphics.plot_regress_exog(model,'ADMNSTRN',fig = fig)
```

```python
fig = plt.figure(figsize = (15,8))
sm.graphics.plot_regress_exog(model,'MRKTNGSPND',fig = fig)
```

MODEL DELETION DIAGNOSTICS (CHECKING OUTLIERS OR INFLUENCERS)

THERE ARE TWO TYPES OF TECHNIQUES:
    
    COOK'S DISTANCE
    
    LEVERAGE VALUE

```python
# 1.COOK'S DISTANCE (if cook's diastance > 1, then its an outlier)
(c,_) = model.get_influence().cooks_distance
c
```

```python
# plotting influencers using stem plots
fig = plt.figure(figsize = (15,8))
plt.stem(np.arange(len(startup)),np.round(c,5))
plt.xlabel('ROW INDEX')
plt.ylabel("COOK'S DISTANCE")
plt.show()
```

```python
# index and value of influence (c>0.5)
np.argmax(c) , np.max(c)
```

```python
# 2. LEVERAGE CUTOFF (points beyond leverage cutoff are influencers)
influence_plot(model)
plt.show()
```

```python
# leverage cut-off value = 3 * (k+1)/n ; k = no. of features/columns and n = no. of datapoints
k = startup.shape[1]
n = startup.shape[0]
leverage_cutoff = 3*(k+1)/n
leverage_cutoff
```
